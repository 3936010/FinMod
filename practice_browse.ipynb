```python
from langchain_openai import ChatOpenAI
from langchain_google_genai import ChatGoogleGenerativeAI
from browser_use import Agent, Browser, ChatBrowserUse
import asyncio
from dotenv import load_dotenv
import os
load_dotenv()

openai_api_key = os.getenv("OPENAI_API_KEY")
```

```python
async def example():
    browser = Browser(
        # use_cloud=True,  # Uncomment to use a stealth browser on Browser Use Cloud
    )

    llm = ChatBrowserUse()

    task = """
    Go to website 'http://10.10.113.400:8001' 
    1. create an account with a random username and password
    2. login with the created account
    3. there is a button called '+ Yangi' on the main page after login, click it
    4. Ask the support agent 'Windows 11 ni qanday o'rnatish mumkin?'.
    """

    agent = Agent(
        task=task,
        llm=llm,
        browser=browser,
    )

    history = await agent.run()
    return history


history = asyncio.run(example())
```
